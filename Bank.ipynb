```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import GridSearchCV 
import math
import csv
from sklearn.model_selection import train_test_split
import collections
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
datasetFields = []

bankCSV = "bank-full.csv"

with open(bankCSV,'rt') as bcsv:
    reader = csv.reader(bcsv, delimiter=';')
    for row in reader:
        datasetFields.append(row)
print(datasetFields[:][1])
```

```python
bank_df = pd.DataFrame(datasetFields[1:], columns = datasetFields[0])
#bank_df
```

```python
bank_df.describe()
```

```python
#Converting to numeric labels

#attributes = []

#for column in bank_df.columns.tolist():
  #  val = set(bank_df[column])
  #  attributes.append(list(val))
  #  val_map = {val:i for val,i in zip(attributes[-1],range(len(val)))}
  #  bank_df[column] = bank_df[column].replace(val_map)
```

```python
# One hot encoding of columns
columns_to_encode = bank_df.columns

for name in columns_to_encode:
    one_hot = pd.get_dummies(bank_df[name])
    bank_df = bank_df.drop(name,axis = 1)
    bank_df = pd.concat([bank_df,one_hot],axis=1,sort=False)
    bank_df = bank_df.rename(columns={'yes': name})
    #bank_df = bank_df.drop(columns=['no'])
#bank_df 
```

```python
y = bank_df.y.as_matrix()
bank_df.drop('y',axis=1,inplace=True)

X = pd.DataFrame()
X = bank_df
X.shape
```

```python
knn_parameters = {"neighbors": [1,10,20,30,40,50,100,150]}

for i in range(1,3):

    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

    #knn
    knn_classifier = KNeighborsClassifier()  
    classifier = GridSearchCV(knn_classifier,knn_parameters,cv=3,n_jobs=4)
    knn_classifier.fit(x_train, y_train)  
    
    
    
print ("For 80-20 train-test split, the test score for KNN classifier is: ", knn_classifier.score(x_test,y_test))
```

# 3) Tic-Tac-Toe

```python
datasetTTT = []

TTTData = "tic-tac-toe.data"

with open(TTTData,'rt') as ttt:
    
    for line in ttt:
        line = line.translate({ord(c): None for c in '\n'})
        line = line.replace("positive", "1")
        line = line.replace("negative", "0")
        splitter = line.split(',')
        datasetTTT.append(splitter)

datasetTTT.pop(-1)
datasetTTT=([['TopLeft','TopMid','TopRight','MidLeft','MidMid','MidRight','BotLeft','BotMid','BotRight','WinLoss']])+datasetTTT
print(datasetTTT[:][0])

ttt_df = pd.DataFrame(datasetTTT[1:], columns = datasetTTT[0])
ttt_df
```

```python
ttt_df.describe()
```

```python
columns_to_encode = ttt_df.columns
for name in columns_to_encode:
    one_hot = pd.get_dummies(ttt_df[name])
    #ttt_df = ttt_df.drop(name,axis = 1)
    ttt_df = pd.concat([ttt_df,one_hot],axis=1,sort=False)
    ttt_df = ttt_df.rename(columns={'yes': name})
ttt_df 
```

```python
y = ttt_df.WinLoss.as_matrix()
ttt_df.drop('WinLoss',axis=1,inplace=True)

X = pd.DataFrame()
X = ttt_df
X.shape
```

```python
X.drop('TopLeft',axis=1,inplace=True)
X.drop('TopMid',axis=1,inplace=True)
X.drop('TopRight',axis=1,inplace=True)
X.drop('MidLeft',axis=1,inplace=True)
X.drop('MidMid',axis=1,inplace=True)
X.drop('MidRight',axis=1,inplace=True)
X.drop('BotLeft',axis=1,inplace=True)
X.drop('BotMid',axis=1,inplace=True)
X.drop('BotRight',axis=1,inplace=True)
```

```python
X.drop(columns=['0','1'],inplace=True)
```

```python
X.shape
```

```python
knn_parameters = {"neighbors": [1,10,20,30,40,50,100,150]}

for i in range(1,3):

    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

    #knn
    knn_classifier = KNeighborsClassifier()  
    classifier = GridSearchCV(knn_classifier,knn_parameters,cv=3,n_jobs=4)
    knn_classifier.fit(x_train, y_train)  
    
    
    
print ("For 80-20 train-test split, the test score for KNN classifier is: ", knn_classifier.score(x_test,y_test))
```
